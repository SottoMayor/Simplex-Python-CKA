```python
from scipy.optimize import linprog
import numpy as np
import matplotlib.pyplot as plt

# Definir as equações
def eq1(x):
    return (150 - 3*x)/6

def eq2(x):
    return 22 - 2*x

def eq3(x):
    return 27.5 - x

# Definir os valores de x
x = np.linspace(0, 50, 100)

# Plotar cada desigualdade separadamente
plt.plot(x, eq1(x), label='3x+6y<=150')
plt.plot(x, eq2(x), label='x+0.5y<=22')
plt.plot(x, eq3(x), label='x+y<=27.5')

# Adicionar as restrições x>=0 e y>=0
plt.axvline(0, color='black')
plt.axhline(0, color='black')

# Adicionar legendas e título
plt.legend()
plt.title('Desigualdades separadas')
plt.show()



# Criar a matriz de coeficientes e o vetor de constantes
A = np.array([[3, 6], [1, 0.5], [1, 1], [-1, 0], [0, -1]])
b = np.array([150, 22, 27.5, 0, 0])

# Encontrar a interseção das desigualdades
P = polytope(A, b)
V = P.vertices

# Plotar as desigualdades e a interseção
plt.plot(x, eq1(x), label='3x+6y<=150')
plt.plot(x, eq2(x), label='x+0.5y<=22')
plt.plot(x, eq3(x), label='x+y<=27.5')
plt.fill(V[:,0], V[:,1], 'green', alpha=0.2)

# Adicionar as restrições x>=0 e y>=0
plt.axvline(0, color='black')
plt.axhline(0, color='black')

# Adicionar legendas e título
plt.legend()
plt.title('Desigualdades juntas')
plt.show()



'''
Forma geral:
    min c'x
    s.a: Ax <= b
    
OBS: min (c'x) é igual max (-c'x)
'''

# coeficientes da função objetivo
c = [-20, -30]

# matriz de coeficientes das restrições
A = [[3, 6], [1, 0.5], [1, 1]]

# vetor de termos independentes das restrições
b = [150, 22, 27.5]

# limites das variáveis (A e B)
x0_bounds = (0, None)
x1_bounds = (0, None)

# resolvendo o problema de programação linear
res = linprog(c, A_ub=A, b_ub=b, bounds=[x0_bounds, x1_bounds], method='simplex')

# imprimindo informações relacionados à implementação
print('Tudo certo na execução do programa.' if res.success else 'Opa! Algo deu errado. Tente novamente.' )
print('Quantidade de interações para chegar no ótimo: ', res.nit)

print('\n')

# imprimindo a solução ótima
print('Quantidade de bolos do tipo A: ', round(res.x[0], 2))
print('Quantidade de bolos do tipo B: ', round(res.x[1], 2))
print('Lucro máximo: R$', round(-res.fun))
```
